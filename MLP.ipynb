```python
from torch.utils.data import DataLoader, Dataset
import random as random
from torch import cat, nn, mm, transpose
import torch.nn.functional as F
```

```python
class MLP(nn.Module):
    def __init__(self, num_users, num_items, latent_dim):
        super().__init__()
        # embedding layers
        self.embedding_user_mlp = nn.Embedding(num_users, latent_dim)
        self.embedding_item_mlp = nn.Embedding(num_items, latent_dim)
        
        # define four layers
        # first layer is the embedding layer
        self.fc1 = nn.Linear(16, 32)
        self.fc2 = nn.Linear(32,16)
        self.fc3 = nn.Linear(16,8)
        self.fc4 = nn.Linear(8,1)
        self.dropout = nn.Dropout(p=0.5)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

        
    def forward(self, users, items):
        embedding_user = self.embedding_user_mlp(users)
        embedding_item = self.embedding_item_mlp(items)
        # concatenate embedding vectors 
        x = cat([embedding_user, embedding_item], dim=1)
        x = self.dropout(self.relu(self.fc1(x)))
        x = self.dropout(self.relu(self.fc2(x)))
        x = self.dropout(self.relu(self.fc3(x)))
        # output layer doesn't need relu
        x = self.fc4(x)
        
        return self.sigmoid(x) 

        
```
