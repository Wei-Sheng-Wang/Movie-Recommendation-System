```python
from torch.utils.data import DataLoader, Dataset
import random as random
from torch import cat, nn, mul, transpose
```

```python
class GMF(nn.Module):
    def __init__(self, num_users, num_items, latent_dim):
        super(GMF, self).__init__()
        self.num_users = num_users
        self.num_items = num_items
        self.latent_dim_mf = latent_dim
        
        # layer embeddings
        self.embedding_user_mf = nn.Embedding(self.num_users, self.latent_dim_mf)
        self.embedding_item_mf = nn.Embedding(self.num_items, self.latent_dim_mf)

        self.embedding_layer = nn.Linear(latent_dim, 1)
        self.sigmoid = nn.Sigmoid()
        
        
    def forward(self, users, items):
        # embedding vectors
        embedding_user_mf = self.embedding_user_mf(users)
        embedding_item_mf = self.embedding_item_mf(items)

        # element wise product
        mf_matrix = mul(embedding_user_mf, embedding_item_mf)
        x = self.embedding_layer(mf_matrix)
        return self.sigmoid(x) 
    
```
